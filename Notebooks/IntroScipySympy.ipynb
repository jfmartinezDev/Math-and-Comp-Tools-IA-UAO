$f(x) = \sqrt{x-1}$

Dominio: $x\geq 1$

Rango: $y\geq 0$

```
x = np.linspace(1,100,1000)
fx = np.sqrt(x-1)
plt.plot(x,fx)
plt.show()
```

\begin{equation*}
f(x) = 
\begin{cases} 
1-x, & \mbox{if } x< 1 \\ 
\sqrt{x-1}, & \mbox{if } x\geq 1 
\end{cases}
\end{equation*}

```
x1 = np.linspace(-10,0.99,1000)
x2 = np.linspace(1,10,1000)
fx1 = 1-x1
fx2 = np.sqrt(x2-1)
plt.plot(x1,fx1,x2,fx2)

plt.show()
```

```
import sympy as sp
from sympy.plotting import plot
#from sympy import *
```

```
x= sp.Symbol('x')
print(sp.diff(2*x**4 + 7*x**3 + 6*x**2 + 2*x +15))
p1=sp.plot(x**2,show=False)
p2=sp.plot(2*x,show=False)
p1.append(p2[0])
p1.show()
```

```
x = sp.Symbol('x')
f = 2*x**4 + 7*x**3 + 6*x**2 + 2*x +15
f_prime = f.diff(x) 
```

```
func = sp.lambdify(x,f)
f_prime = sp.lambdify(x,f_prime)

print(func(1))
print(f_prime(1))
```

Derivada de una funcion en un punto

```
from scipy.misc import derivative

def f(x):

    return 2*x**4 + 7*x**3 + 6*x**2 + 2*x +15

derivative(f, 1.0, dx=1e-6)
```

Ejemplos Proyecto

Ejemplo 1

```
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

Rangox1=np.arange(-10,10,0.1)
Rangox2=np.arange(-4,4,0.1)
X1,X2=np.meshgrid(Rangox1,Rangox2)
Z=0.5*(X1**2+10*X2**2)

# Plot the surface.
surf = ax.plot_surface(X1, X2, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
```

Algoritmo del gradiente: Ejemplo 1 Minimizar la funcion:

$$
0.5(x_1^2+10x_2^2)$$

```
alpha=0.15 #alpha debe estar entre [0,1]

X1min=[]
X2min=[]

X1min.append(10) #condicion inicial
X2min.append(4) #condicion inicial

fig2, ax = plt.subplots()

for i in np.arange(0,100):
  
  X1min.append(X1min[i]+(alpha*(-X1min[i])))
  X2min.append(X2min[i]+(alpha*(-10*X2min[i])))

plt.plot(X1min,X2min)

ax.contour(X1, X2, Z)
plt.show()
```

Ejemplo 2

```
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

Rangox1=np.arange(-100,100,1)
Rangox2=np.arange(-50,50,1)
X1,X2=np.meshgrid(Rangox1,Rangox2)
Z=X1**2 - 24*X1 + X2**2 - 10*X2

# Plot the surface.
surf = ax.plot_surface(X1, X2, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
```

Algoritmo del gradiente: Ejemplo 2 Minimizar la funcion:

$$
x_1^2 − 24x_1 + x_2^2− 10x_2$$

```
alpha=0.1 #alpha debe estar entre [0,1]

X1min=[]
X2min=[]

X1min.append(-95) #condicion inicial
X2min.append(49) #condicion inicial

fig2, ax = plt.subplots()

for i in np.arange(0,100):
  
  X1min.append(X1min[i]+(alpha*(-2*X1min[i]-24)))
  X2min.append(X2min[i]+(alpha*(-2*X2min[i]-10)))

plt.plot(X1min,X2min)
ax.contour(X1, X2, Z)

plt.show()
```
