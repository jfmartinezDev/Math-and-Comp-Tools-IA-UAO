```
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import sympy as sp
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
```

```
x = np.linspace(-10, 10, 50)

y = x

dy = np.ones([30])

plt.plot(x, y, "r--", label="y=x" )
plt.plot(dy,"g-", label="y'= 1" )

plt.title("FUNCIONES Y SUS DERIVADAS")

plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```
x = np.linspace(-10,10,50)

y = x**2

yPrime = 2*x

plt.plot(x, y, "r--", label="y = x**2" )
plt.plot(x, yPrime,"g-", label="y' = 2x" )

plt.title("FUNCIONES Y SUS DERIVADAS")

plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```
x = np.linspace(-20, 20, 50)

p = 2*(x**4) + 7*(x**3) + 6*(x**2) + 2*x

pPrime = 8*(x**3) + 21*(x**2) + 12*(x) + 2

plt.plot(x, p, "r--", label="p = 2*(x**4) + 7*(x**3) + 6*(x**2) + 2*x")
plt.plot(pPrime,"g-", label="p' = 8*(x**3) + 21*(x**2) + 12*(x) + 2" )

plt.title("FUNCIONES Y SUS DERIVADAS")

plt.legend(loc="upper right")
plt.grid(True)
plt.show()
```

```
from sympy import symbols
from numpy import linspace
from sympy import lambdify
import matplotlib.pyplot as mpl
x = symbols('x')
f= 2* x**4 + 7 * x**3 + 6 * x**2 + 2 * x + 15
f_prime = sp.diff(f)
func = lambdify(x,f, modules=['numpy'])
f_prime = lambdify(x,f_prime, modules=['numpy'])
x_vals = linspace(-10, 10, 100)
y_vals = func(x_vals)
y_prime_vals = f_prime(x_vals)
mpl.plot(x_vals, y_vals)
mpl.plot(x_vals, y_prime_vals)
plt.grid(True)
mpl.show()
```

```
from mpl_toolkits.mplot3d import Axes3D

figure = plt.figure(1, figsize = (12, 12))
subplot3d = figure.add_subplot(111, projection='3d')

x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 50)
X, Y = np.meshgrid(x, y)

Z1 = (X**2) + (Y**2)
surface1 = subplot3d.plot_surface(X, Y, Z1, rstride=1, cstride=1, color="m", linewidth=0.1, alpha=0.8)


subplot3d.set_xlabel('X')
subplot3d.set_ylabel('Y')
subplot3d.set_zlabel('Z')
plt.legend(loc="lower right")

plt.show()
```
