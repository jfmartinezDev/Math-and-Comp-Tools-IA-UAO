### PUNTO 4: 
Escriba un programa que reciba como entrada las longitudes de los dos catetos A y B de un triángulo
rectángulo, y que entregue como salida el largo de la hipotenusa C del triángulo, dado por el teorema de Pitágoras: 
C2 = A2 + B2


```
import math
print("Hola, este es un programa escrito en Python con Google COLAB\n que calcula a hipotenusa 'C' de un triángulo rectángulo \n a partir de los valores de los catetos 'A' y 'B' \n solicitados por teclado al usuario:\n")
A = float(input("Por favor digite la medida de la longitud del cateto A: "))
print("El valor digitado para el cateto A es: ", A)
B = float(input("Por favor digite la medida de la longitud del cateto B: "))
print("El valor digitado para el cateto B es: ", B)
def pythagoras(ca, cb):
  H = math.sqrt((ca**2)+(cb**2))
  return H
C = pythagoras(A, B)
print("El valor de la hipotenusa es: ", C)
```

### PUNTO 5: 
Implemente un programa que reciba como entrada el radio de un círculo y entregue como salida su perímetro y su área.


```
import math
print("Hola, este es un programa escrito en Python con Google COLAB\n que retorna el valor del área y el perímetro de un círculo \n a partir del valor del radio \n solicitado por teclado al usuario:\n")
r = float(input("Por favor digite la medida la longitud del radio: "))
print("El valor digitado para el radio es: ", r)
def area_and_perimeter(R):
  circle = [None, None] #circle=[]
  circle[0] = (math.pi)*(R**2)
  circle[1] = 2*(math.pi)*(R)
  return circle
ans = area_and_perimeter(r)
print("El valor del área del círculo es: ", ans[0])
print("El valor del perímetro del círculo es: ", ans[1])
```

### PUNTO 1: 
Escriba un programa que reciba como entrada dos números, y los muestre ordenados de menor a mayor.

```
print("Hola, este es un programa escrito en Python con Google COLAB\n que retorna dos números ordenados de menor a mayor:\n")
n1 = float(input("Por favor digite el primer número: "))
print("El valor digitado para el primer número: ", n1)
n2 = float(input("Por favor digite el segundo número: "))
print("El valor digitado para el segundo número es: ", n2)
def menor_y_mayor(a, b):
  order = [None, None]
  if a<b:
    order[0]= a
    order[1]= b
    return order
  elif a>b:
      order[0]= b #append
      order[1]= a
      return order
  else:
    return False

ans = menor_y_mayor(n1, n2)
if ans==False:
   print("Los números son iguales")
else:
  print("Los números ordenados de menor a mayor: ", ans[0], ans[1])
```

###PUNTO 2:
Realice un programa que simule una calculadora básica, este puede realizar operación de suma, resta, multiplicación y división. 
El programa debe recibir como entrada 2 números reales y un operador, que puede ser +, −, ∗, /. La salida del programa debe ser el resultado de la operación.

```
print("Hola, Bienvenido a CALCULADORA:\n")
print("Este programa permite calcular cualquiera de las 4 operaciones básicas entre 2 números ingresados por el usuario:\n")
n1 = float(input("Por favor digite el primer número: "))
print("El valor digitado para el primer número: ", n1)
op = input("Por favor digite el simbolo de la operación que desea,\n bien sea + para suma, - para resta, * para multiplicación, o / para división : ")
print("La operación seleccionada es: ", op)
n2 = float(input("Por favor digite el segundo número: "))
print("El valor digitado para el segundo número es: ", n2)

def calculator(a, b, oper):
  ans=None
  if oper=="+":
    ans=a+b
  elif oper=="-":
    ans=a-b
  if oper=="*":
    ans=a*b
  if oper=="/":
    ans=a/b
  return ans

print("El resultado de la operación es: ", calculator(n1, n2, op))

  
```

##PUNTO 2:
Escriba un programa que pida al usuario dos números enteros, y luego entregue la suma de todos los números que están entre ellos. Por ejemplo, si los números son 1 y 7, debe entregar como resultado: 2 + 3 + 4 + 5 + 6 = 20.

```
print("Bienvenido, este programa recibe 2 números enteros digitados por el usuario y \n calcula el valor de todos los números enteros entre esos 2 números, excluyendolos :\n")
num1 = int(input("Por favor digite el primer número: "))
print("El primer número ingresado es: ", num1)
num2 = int(input("Por favor digite el segundo número: "))
print("El segundo número ingresado es: ", num2)
print("\n")

def sumbetween(a, b):
  sum=0
  for i in range(a+1, b, 1):
    sum=sum+i
  print(sum, "\n")

sumbetween(num1, num2)
```
